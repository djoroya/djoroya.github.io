```python
import svgwrite
import math

# Crear el archivo SVG
dwg = svgwrite.Drawing("logo_dj_with_hexagons.svg", profile="tiny")



# Color y estilo black
color_azul = "white"
# Parámetros de letra
font_size = 70
x_position_d = 30
x_position_j = 70
xmu = x_position_j
y_position = 100
ymu =y_position*0.75
# Agregar la letra 'D' al SVG

def draw_rectangle(dwg, xmu, ymu, hex_size, color_azul,fill="none"):
    dwg.add(dwg.rect(
        insert=(xmu-0.5*hex_size, ymu-0.5*hex_size),
        size=(hex_size, hex_size),
        fill=fill,
        stroke=color_azul,
        stroke_width=6
    ))
# Parámetros de la malla de hexágonos
hex_size = 15  # Tamaño del hexágono
hex_spacing = 1.5*hex_size * math.sqrt(3)  # Espaciado para evitar superposición
rows = 5  # Número de filas de hexágonos
cols = 6  # Número de columnas de hexágonos
x_offset = 1.5*hex_spacing  # Ajusta la posición de la malla horizontalmente
y_offset = 1.5*hex_spacing  # Ajusta la posición de la malla verticalmente
# se dibujan los rectangulos
draw_rectangle(dwg, xmu, ymu, 8.5*hex_size, color_azul,fill="black")
# draw_rectangle(dwg, xmu, ymu, 7.5*hex_size, color_azul,fill="white")


dwg.add(dwg.text(
    "D",
    insert=(x_position_d, y_position),
    font_size=font_size,
    font_family="Arial",
    fill=color_azul,
    font_weight="bold"
))

# Agregar la letra 'J' al SVG
dwg.add(dwg.text(
    "J",
    insert=(x_position_j, y_position),
    font_size=font_size,
    font_family="Arial",
    fill=color_azul,
    font_weight="bold"
))

# Función para dibujar un hexágono
def draw_circle(dwg, x, y, r, color):
    dwg.add(dwg.circle(
        center=(x, y),
        r=r,
        fill=color
    ))



# Generar la malla de hexágonos detrás de las letras
#  plot
draw_circle(dwg, xmu, ymu, hex_size, color_azul)
# las cuatro esquinas
for i in [-1,1]:
    for j in [-1,1]:
        draw_circle(dwg, 
                    xmu + i*hex_spacing, 
                    ymu + j*hex_spacing, 
                    hex_size, color_azul)



# Guardar el archivo SVG
dwg.save()
```

```python
import svgwrite
import math

# Crear el archivo SVG
dwg = svgwrite.Drawing("logo_dj_with_hexagons_modified.svg", profile="tiny", size=("200px", "200px"))

# Color y estilo: blanco sobre fondo negro
color_azul = "white"
background_color = "black"

# Fondo negro rectangular
dwg.add(dwg.rect(
    insert=(0, 0),
    size=("100%", "100%"),
    fill=background_color
))

# Parámetros de letra
font_size = 70
x_position_d = 40
x_position_j = 85
y_position = 110

# Función para dibujar un hexágono
def draw_circle(dwg, x, y, r, color):
    dwg.add(dwg.circle(
        center=(x, y),
        r=r,
        fill=color
    ))

# Dibujar el rectángulo central detrás de las letras "DJ"
rect_size = 8.5 * 15  # Ajusta el tamaño en función de la malla
draw_rectangle(dwg, x_position_j, y_position * 0.75, rect_size, color_azul, fill=background_color)

# Agregar la letra 'D' al SVG
dwg.add(dwg.text(
    "D",
    insert=(x_position_d, y_position),
    font_size=font_size,
    font_family="Arial",
    fill=color_azul,
    font_weight="bold"
))

# Agregar la letra 'J' al SVG
dwg.add(dwg.text(
    "J",
    insert=(x_position_j, y_position),
    font_size=font_size,
    font_family="Arial",
    fill=color_azul,
    font_weight="bold"
))

# Parámetros de la malla de círculos
hex_size = 15
hex_spacing = 1.5 * hex_size * math.sqrt(3)
xmu = (x_position_d + x_position_j) / 2  # Centro entre las letras
ymu = y_position * 0.75

# Dibujar círculo central y los círculos en las cuatro esquinas
draw_circle(dwg, xmu, ymu, hex_size, color_azul)
for i in [-1, 1]:
    for j in [-1, 1]:
        draw_circle(dwg, xmu + i * hex_spacing, ymu + j * hex_spacing, hex_size, color_azul)

# Guardar el archivo SVG
dwg.save()
```
